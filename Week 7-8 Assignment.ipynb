# <h1><center>Data Cleaning and Transformations</center></h1>

The goal of this assignment is to clean and transform two datasets.  The first one is the candy data set and the second is the met data set. The candy dataset is survey data of candy. The met dataset is a collection of artwork held at the Metropolitan Museum of Art. With in the analysis I will perform several tasks. The tasks include Filter out missing data, Transform data using either mapping or a function, Create hierarchical index, Pivot the data, Grouping with Functions, Split/Apply/Combine, Generate date range, and Convert timestamps to periods and back.

```python
# This code imports the necessary libraries
import pandas as pd
import warnings
```

```python
# This code suppresses minor warnings
warnings.simplefilter('ignore')
```

```python
# This code uploads the candy dataset and displays the first five rows
candy = pd.read_excel(r'candyhierarchy2017.xlsx')
candy.head()
```

```python
# This code imports the met data set and displays the first 5 rows
met = pd.read_csv('MetObjects.txt', sep=",")
met.head()
```

1. Filter out missing data

```python
# This code displays the types of data in the candy data set
candy.dtypes
```

```python
# This code gives the dimension for the candy data
candy.shape
```

```python
# This code displays the sum of the null values
candy.isnull().sum()
```

```python
# This code fills the null values with a zero and displays the sum of any null values
candy_1 = candy.fillna(0)
candy_1.isnull().sum()
```

```python
# This code displays the first five rows after replacing null values
candy_1.head()
```

2. Transform data using either mapping or a function

```python
# This code maps to the department column and adds "This is the department of" in front of the department title
met_1 = met["Department"].map("This is the department of {}".format)
met_1.head()
```

3. Create hierarchical index

```python
# this code displays the range of the candy index
candy.index
```

```python
# This code creates an index hierarchy for the candy data
candy_2 = candy_1.set_index(["Q4: COUNTRY", "Q5: STATE, PROVINCE, COUNTY, ETC", "Q2: GENDER"])
candy_2.head()
```

4. Pivot the data

```python
# This code displays the first five rows of the met data
met.head()
```

```python
# This code pivots the met data 
pivot_met = met.pivot(index='AccessionYear', columns='Object ID', values='Object Number' )
pivot_met.tail()
```

5. Grouping with Functions

```python
# This code converts the accession year column from an object to a date
met['AccessionYear']= pd.to_datetime(met['AccessionYear'])
```

```python
# This code converts the object number column from an object to a number
met['Object Number'] = pd.to_numeric(met['Object Number'], errors='coerce')
```

```python
# This code creates a group by using accession year and object number and displays the first five rows
met_group = met.groupby('AccessionYear')['Object Number'].mean()
met_group.head()
```

6. Split/Apply/Combine

```python
# This code renames the object name column 
met = met.rename(columns={'Object Name' : "Object_Name"})
```

```python
# this code displays the unique values in the column Object_Name
met["Object_Name"].unique
```

```python
# This code creates a new data frame grouped by object name
met_2 = met.groupby('Object_Name')
met_2
```

```python
# This code creates a for loop to cycle through the object name in the met_2 data frame
for Object_Name, Object_Name_met in met_2:
    print(Object_Name)
    print(Object_Name_met)
```

```python
# This code creates a data frame of the prints in the met_2 data frame
met_2.get_group('Print')
```

7. Generate date range

```python
# This code displays the dimension of the candy_1 data frame
candy_1.shape
```

```python
# This code creates a date range for the candy_1 data set
candy_1["Date"] = pd.date_range(start="2020-01-01", periods=2460, freq="D")
candy_1.head()
```

8. Convert timestamps to periods and back

```python
# This converts the accession year column from an object to a time stamp
met["AccessionYear"]= pd.to_datetime(met["AccessionYear"])
```

```python
# This code converts the time stamp into a period and displays the result
met["Accession_Period"]= met["AccessionYear"].dt.to_period("M")
met["Accession_Period"]
```

```python
# This code converts the period back into a time stamp and displays the result
met["AccessionYear_converted"] = met["Accession_Period"].dt.to_timestamp()
met["AccessionYear_converted"]
```
